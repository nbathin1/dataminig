```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from tqdm import tqdm
def randomforest_batch(df,batch_size,input_features,input_size):
    #batch_size = 1600
    num_samples = len(df)
    num_samples = len(df)
    indices = np.arange(num_samples)
    num_steps_per_epoch = num_samples / batch_size 

    for start_idx in tqdm(range(0, num_samples, batch_size),desc="Batch_processed"):
        end_idx = min(start_idx + batch_size, num_samples)
        excerpt = indices[start_idx:end_idx]
        X = np.array([np.reshape(x, (input_features*input_size,)) for x in df.iloc[excerpt]['x']])
        Y = df.iloc[excerpt]['y']
        x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
        rf = RandomForestClassifier(n_estimators=100)
        rf.fit(x_train, y_train)
        y_pred = rf.predict(x_test)
        accuracy = accuracy_score(y_test, y_pred)
        correctly_predicted = sum((y_pred == 2) & (y_test == 2))
        num_correct = np.sum((y_test == 2))
        accuracy_list.append(accuracy)
        correctly_predicted_list.append(correctly_predicted)
        num_correct_list.append(num_correct)
        #progress_bar.update(1)
        
        
    total_correctly_predicted = sum(correctly_predicted_list)
    total_num_correct = sum(num_correct_list)
    total_accuracy = np.mean(accuracy_list)
    total_precision = total_correctly_predicted / total_num_correct
    total_positive_percentage = np.sum(total_num_correct)/ (len(df)*0.2)
    return total_accuracy, total_precision, total_positive_percentage
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

class NeuralNetwork(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(NeuralNetwork, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size)
        self.activation = nn.ReLU()
        self.fc2 = nn.Linear(hidden_size, output_size)
        self.regularization = nn.L1Loss()
        
    def forward(self, x):
        out = self.fc1(x)
        out = self.activation(out)
        out = self.fc2(out)
        return out



def train_nn(model, df, batch_size, num_epochs, input_size, hidden_size, output_size, optimizer, lr,a,b):
    # Initialize the neural network
    net = model(input_size, hidden_size, output_size)
    
    # Define the loss function and optimizer
    criterion = nn.CrossEntropyLoss()
    if optimizer == 'adam':
        optimizer = optim.Adam(net.parameters(), lr=lr)
    elif optimizer == 'sgd':
        optimizer = optim.SGD(net.parameters(), lr=lr, momentum=0.9)
    else:
        raise ValueError("Invalid optimizer")
    
    num_samples = len(df)
    indices = np.arange(num_samples)
    np.random.shuffle(indices)
    accuracy_list = []
    correctly_predicted_list = []
    num_correct_list = []
    for epoch in range(num_epochs):
        for start_idx in tqdm(range(0, num_samples, batch_size), desc="Batch_processed"):
            end_idx = min(start_idx + batch_size, num_samples)
            excerpt = indices[start_idx:end_idx]
            X = np.array([np.reshape(x, (a*b,)) for x in df.iloc[excerpt]['x']])
            Y = df.iloc[excerpt]['y'].values
            X = torch.Tensor(X)
            Y = torch.LongTensor(Y)
            x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
            net.train()
            optimizer.zero_grad()
            y_pred = net(x_train)
            loss = criterion(y_pred, y_train)
            loss.backward()
            optimizer.step()
            net.eval()
            y_pred_test = net(x_test)
            y_pred_test = torch.argmax(y_pred_test, dim=1)
            if epoch==num_epochs-1:
                accuracy = accuracy_score(y_test, y_pred_test)
                correctly_predicted = sum((y_pred_test == 2) & (y_test == 2))
                num_correct = sum((y_test == 2))
                accuracy_list.append(accuracy)
                correctly_predicted_list.append(correctly_predicted)
                num_correct_list.append(num_correct)
    
    total_correctly_predicted = sum(correctly_predicted_list)
    total_num_correct = sum(num_correct_list)
    total_accuracy = np.mean(accuracy_list)
    total_precision = total_correctly_predicted / total_num_correct
    total_positive_percentage = total_num_correct/ (len(df)*0.2)
    return total_accuracy, total_precision, total_positive_percentage
```

```python
class Model1(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(Model1, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size)
        self.relu1 = nn.ReLU()
        self.fc2 = nn.Linear(hidden_size, output_size)
        self.dropout = nn.Dropout(p=0.2)
        self.relu2 = nn.ReLU()

    def forward(self, x):
        out = self.fc1(x)
        out = self.relu1(out)
        out = self.dropout(out)
        out = self.fc2(out)
        out = self.relu2(out)
        return out


class Model2(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(Model2, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size)
        self.batch_norm1 = nn.BatchNorm1d(hidden_size)
        self.relu1 = nn.ReLU()
        self.fc2 = nn.Linear(hidden_size, output_size)
        self.dropout = nn.Dropout(p=0.3)
        self.relu2 = nn.ReLU()

    def forward(self, x):
        out = self.fc1(x)
        out = self.batch_norm1(out)
        out = self.relu1(out)
        out = self.dropout(out)
        out = self.fc2(out)
        out = self.relu2(out)
        return out


class Model3(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(Model3, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size)
        self.relu1 = nn.ReLU()
        self.fc2 = nn.Linear(hidden_size, hidden_size)
        self.relu2 = nn.ReLU()
        self.fc3 = nn.Linear(hidden_size, output_size)
        self.dropout1 = nn.Dropout(p=0.2)
        self.dropout2 = nn.Dropout(p=0.3)

    def forward(self, x):
        out = self.fc1(x)
        out = self.relu1(out)
        out = self.dropout1(out)
        out = self.fc2(out)
        out = self.relu2(out)
        out = self.dropout2(out)
        out = self.fc3(out)
        return out
```

```python
def train_nn1(model, df, batch_size, num_epochs, optimizer, lr,a,b):
    # Initialize the neural network
    net = model()
    
    # Define the loss function and optimizer
    criterion = nn.CrossEntropyLoss()
    if optimizer == 'adam':
        optimizer = optim.Adam(net.parameters(), lr=lr)
    elif optimizer == 'sgd':
        optimizer = optim.SGD(net.parameters(), lr=lr, momentum=0.9)
    else:
        raise ValueError("Invalid optimizer")
    
    num_samples = len(df)
    indices = np.arange(num_samples)
    np.random.shuffle(indices)
    accuracy_list = []
    correctly_predicted_list = []
    num_correct_list = []
    for epoch in range(num_epochs):
        for start_idx in tqdm(range(0, num_samples, batch_size), desc="Batch_processed"):
            end_idx = min(start_idx + batch_size, num_samples)
            excerpt = indices[start_idx:end_idx]
            X = np.array([np.reshape(x, (a*b,)) for x in df.iloc[excerpt]['x']])
            Y = df.iloc[excerpt]['y'].values
            X = torch.Tensor(X)
            Y = torch.LongTensor(Y)
            x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
            net.train()
            optimizer.zero_grad()
            y_pred = net(x_train)
            loss = criterion(y_pred, y_train)
            loss.backward()
            optimizer.step()
            net.eval()
            y_pred_test = net(x_test)
            y_pred_test = torch.argmax(y_pred_test, dim=1)
            if epoch==num_epochs-1:
                accuracy = accuracy_score(y_test, y_pred_test)
                correctly_predicted = sum((y_pred_test == 2) & (y_test == 2))
                num_correct = sum((y_test == 2))
                accuracy_list.append(accuracy)
                correctly_predicted_list.append(correctly_predicted)
                num_correct_list.append(num_correct)
    
    total_correctly_predicted = sum(correctly_predicted_list)
    total_num_correct = sum(num_correct_list)
    total_accuracy = np.mean(accuracy_list)
    total_precision = total_correctly_predicted / total_num_correct
    total_positive_percentage = total_num_correct/ (len(df)*0.2)
    return total_accuracy, total_precision, total_positive_percentage
```

```python
import torch
import torch.nn as nn

class LSTM(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, output_size, dropout=0.0):
        super(LSTM, self).__init__()

        self.hidden_size = hidden_size
        self.num_layers = num_layers

        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True, dropout=dropout)
        self.bn = nn.BatchNorm1d(hidden_size)
        self.fc = nn.Linear(hidden_size, output_size)
        self.dropout = nn.Dropout(p=dropout)
    
    def forward(self, x):
        # Initialize hidden state and cell state with zeros
        h0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(device=x.device)
        c0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(device=x.device)

        # Pass input through LSTM
        out, _ = self.lstm(x, (h0, c0))

        # Apply batch normalization
        out = self.bn(out[:, -1, :])

        # Pass output through fully connected layer
        out = self.fc(out)

        return out
```

```python
def train_lstm(model, df, epochs, batch_size,model_name):
    # Define loss and optimizer
    criterion = nn.CrossEntropyLoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
    num_samples = len(df)
    indices = np.arange(num_samples)
    
    best_metric = float('-inf')
    best_test_accuracy = float('-inf')
    best_test_precision = float('-inf')
    best_num_pred_2 = 0

    for epoch in range(epochs):
        losses = []
        test_accuracies = []
        test_precisions = []
        num_pred_2s = []
        for start_idx in tqdm(range(0, num_samples, batch_size), desc=f"Epoch {epoch+1}/{epochs}"):
            # Get batch of data
            end_idx = min(start_idx + batch_size, num_samples)
            excerpt = indices[start_idx:end_idx]
            X = np.array(df.iloc[excerpt]['x'].tolist())
            Y = df.iloc[excerpt]['y'].values
            X = torch.Tensor(X)
            Y = torch.LongTensor(Y)
            X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

            # Train model
            model.train()
            optimizer.zero_grad()
            outputs = model(X_train)
            loss = criterion(outputs, y_train)
            loss.backward()
            optimizer.step()

            # Evaluate model
            model.eval()
            with torch.no_grad():
                # Test accuracy
                y_pred = torch.argmax(model(X_test), dim=1)
                test_accuracy = accuracy_score(y_test, y_pred)
                test_accuracies.append(test_accuracy)

                # Test precision
                test_precision = precision_score(y_pred,y_test, average = 'macro')
                test_precisions.append(test_precision)

                # Count number of predicted 2s
                num_pred_2 = torch.sum(y_pred == 2)
                num_pred_2s.append(num_pred_2)

                losses.append(loss.item())

        # Calculate mean of metrics for the entire dataset
        mean_loss = np.mean(losses)
        mean_test_accuracy = np.mean(test_accuracies)
        mean_test_precision = np.mean(test_precisions)
        mean_num_pred_2 = np.sum(num_pred_2s)

        # Print mean metrics for the entire dataset
        print(f"Epoch {epoch+1}/{epochs}, Loss: {mean_loss:.4f}, Test Accuracy: {mean_test_accuracy:.4f}, Test Precision: {mean_test_precision:.4f}, Num Predicted 2s: {mean_num_pred_2}")

        # Update best metric if any of the metrics are better than the current best
        if mean_test_accuracy > best_metric:
            best_metric = mean_test_accuracy
            best_test_accuracy = mean_test_accuracy
            best_test_precision = mean_test_precision
            best_num_pred_2 = mean_num_pred_2
            torch.save(model.state_dict(), 'best_model.pt')
        elif mean_test_precision > best_metric:
            best_metric = mean_test_precision
            best_test_accuracy = mean_test_accuracy
            best_test_precision = mean_test_precision
            best_num_pred_2 = mean_num_pred_2
            torch.save(model.state_dict(), 'best_model.pt')
        elif mean_num_pred_2 > best_metric:
            best_metric = mean_num_pred_2
            best_test_accuracy = mean_test_accuracy
            best_test_precision = mean_test_precision
            best_num_pred_2 = mean_num_pred_2
            torch.save(model.state_dict(), 'best_model.pt')
            
    torch.save(model.state_dict(), f"{model_name}.pt")

    return best_test_accuracy, best_test_precision, best_num_pred_2
```

```python
class DeepLSTM(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, output_size):
        super(DeepLSTM, self).__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        self.output_size = output_size
        
        self.lstm1 = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)
        self.bn1 = nn.BatchNorm1d(hidden_size)
        self.dropout1 = nn.Dropout(0.2)
        
        # Add more LSTM layers if needed
        self.lstm_layers = nn.ModuleList()
        self.bn_layers = nn.ModuleList()
        self.dropout_layers = nn.ModuleList()
        for i in range(num_layers - 1):
            self.lstm_layers.append(nn.LSTM(hidden_size, hidden_size, num_layers, batch_first=True))
            self.bn_layers.append(nn.BatchNorm1d(hidden_size))
            self.dropout_layers.append(nn.Dropout(0.2))
        
        self.fc = nn.Linear(hidden_size, output_size)
    
    def forward(self, x):
        # Pass input through first LSTM layer
        out, _ = self.lstm1(x)
        out = out[:, -1, :]  # Only select the last time step
        out = out.view(out.size(0), -1)  # Reshape to (batch_size, hidden_size)
        out = self.bn1(out)
        out = self.dropout1(out)
        
        # Pass output through remaining LSTM layers
        for i in range(self.num_layers - 1):
            out, _ = self.lstm_layers[i](out.unsqueeze(1))
            out = out.squeeze(1)
            out = self.bn_layers[i](out)
            out = self.dropout_layers[i](out)
        
        out = self.fc(out)
        return out
```

```python
class ComplexLSTM(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, output_size, dropout_rate=0.2):
        super(ComplexLSTM, self).__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        self.output_size = output_size
        
        # First LSTM layer
        self.lstm1 = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)
        self.bn1 = nn.BatchNorm1d(hidden_size)
        self.dropout1 = nn.Dropout(dropout_rate)
        
        # Second LSTM layer
        self.lstm2 = nn.LSTM(hidden_size, hidden_size, num_layers, batch_first=True)
        self.bn2 = nn.BatchNorm1d(hidden_size)
        self.dropout2 = nn.Dropout(dropout_rate)
        
        # Third LSTM layer
        self.lstm3 = nn.LSTM(hidden_size, hidden_size, num_layers, batch_first=True)
        self.bn3 = nn.BatchNorm1d(hidden_size)
        self.dropout3 = nn.Dropout(dropout_rate)
        
        # Output layer
        self.fc = nn.Linear(hidden_size, output_size)
    
    def forward(self, x):
        # First LSTM layer
        out, _ = self.lstm1(x)
        out = out[:, -1, :]  # Only select the last time step
        out = out.view(out.size(0), -1)  # Reshape to (batch_size, hidden_size)
        out = self.bn1(out)
        out = self.dropout1(out)
        
        # Second LSTM layer
        out, _ = self.lstm2(out.unsqueeze(1))
        out = out.squeeze(1)
        out = self.bn2(out)
        out = self.dropout2(out)
        
        # Third LSTM layer
        out, _ = self.lstm3(out.unsqueeze(1))
        out = out.squeeze(1)
        out = self.bn3(out)
        out = self.dropout3(out)
        
        # Output layer
        out = self.fc(out)
        return out
```
